```python
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
remove_tv_show = netflix_df[netflix_df["type"] == "TV Show"].index
remove_tv_show
```

```python
netflix_subset = netflix_df.drop(remove_tv_show)
netflix_subset
```

```python
netflix_movies = netflix_subset[
    ["title", "country", "genre", "release_year", "duration"]
]
netflix_movies
```

```python
remove_duration_movie = netflix_movies[netflix_movies["duration"] >= 60].index
remove_duration_movie
```

```python
short_movies = netflix_movies.drop(remove_duration_movie)
short_movies
```

```python
display(short_movies.head(20))
```

```python
colors = []
for index, row in netflix_movies.iterrows():
    genre = row["genre"]
    if genre == "Children":
        colors.append("blue")
    elif genre == "Documentaries":
        colors.append("purple")
    elif genre == "Stand-Up":
        colors.append("orange")
    else:
        colors.append("gray")
```

```python
fig = plt.figure(figsize=(12, 8))
plt.scatter(
    netflix_movies["release_year"], netflix_movies["duration"], c=colors, marker="o"
)
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.show()
```

```python
#create the first of two panels and set current axis
#create the second of two panels and set current axis get current figure information get current axis information
#plot 1:
plt.subplot(1, 2, 1)
plt.plot(netflix_movies["release_year"], netflix_movies["duration"])
plt.title("DURATION")

#plot 2:

plt.subplot(1, 2, 2)
plt.plot(netflix_movies["release_year"], netflix_movies["duration"])
plt.title("Release_year")
plt.suptitle("Netflix")
plt.show()
```

```python
#Creating empty matplotlib figure with four subplots

plt.subplot(3, 2, 1)
plt.title("DURATION")

plt.subplot(3, 2, 2)
plt.title("Release_year")

plt.subplot(3, 2, 5)
plt.title("country")

plt.subplot(3, 2, 6)
plt.title("title")
plt.show()
```

```python
#Create a subplot with 2 rows and 2 columns
plt.subplot(3, 2, 1)
plt.plot(netflix_movies["release_year"], netflix_movies["duration"])
plt.title("DURATION")

plt.subplot(3, 2, 2)
plt.plot(netflix_movies["release_year"], netflix_movies["duration"])
plt.title("Release_year")

plt.subplot(3, 2, 5)
plt.plot(netflix_movies["release_year"], netflix_movies["duration"])
plt.title("country")

plt.subplot(3, 2, 6)
plt.plot(netflix_movies["release_year"], netflix_movies["duration"])
plt.title("title")
plt.show()
```

```python
#Create a subplot with 2 rows and 1 column

plt.subplot(3, 1, 1)
plt.plot(netflix_movies["release_year"], netflix_movies["duration"])
plt.title("DURATION")
plt.subplot(3, 1, 3)
plt.plot(netflix_movies["release_year"], netflix_movies["duration"])
plt.title("Release_year")
plt.show()
```

```python
#Plot the sinusoid function
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 256)
y = np.sin(x)

plt.plot(x, y)
```
